```python
 # Library Import 
from bs4 import BeautifulSoup as soup
from urllib. request import urlopen as uReq
 
```

```python
my_url="https://www.flipkart.com/search?q=iphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
request_1=uReq(my_url) # Requesting for HTML page  
page_html=request_1.read()  #Reading the html page 
   #Closing 
request_1.close()
```

```python
page_soup=soup(page_html,"html.parser")# Formatting 
```

```python
page_soup
```

```python
container=page_soup.findAll("div",{"class":"_1YokD2 _3Mn1Gg"})
```

```python
container
```

```python
print(soup.prettify(container[0]))# prettify data
```

```python
containers=container[0]
```

```python
containers
```

```python
product_name = containers.findAll("div" , {"class":"_4rR01T"})
```

```python
product_name
```

```python
mobile_name=[]
for i in product_name:
    mobile_name.append(i.text)
```

```python
mobile_name
```

```python
product_price = containers.findAll("div" , {"class":"_30jeq3 _1_WHN1"})
```

```python
product_price
```

```python
mobile_price=[]
for i in product_price:
    mobile_price.append(i.text)
```

```python
mobile_price
```

```python
import pandas as pd
```

```python
phones_name_price=pd.DataFrame(list(zip(mobile_name,mobile_price)),columns=['NAME','PRICE'])
phones_name_price.head(7)
```

```python
phones_name_price.to_csv('iPhones_data.csv')
```
